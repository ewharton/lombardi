```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats

from scipy.optimize import minimize
from scipy.misc import factorial

import lombardi
```

```python
mu = 2
mean, var, skew, kurt = scipy.stats.poisson.stats(mu, moments='mvsk')
```

```python
x = np.arange(
    scipy.stats.poisson.ppf(0.01, mu),
    scipy.stats.poisson.ppf(0.99, mu),
)
x = range(10)
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(x, scipy.stats.poisson.pmf(x, mu), label='poisson pmf')
```

```python
data = lombardi.data.player_metric('Tom Brady', 'passing_yds')
x = np.linspace(0, 600, 1000)
x, y_g = lombardi.stats.fit_gamma(x, data, normalize=True)
x, y_p = lombardi.stats.fit_poisson(x, data, normalize=True)

plt.figure(figsize=(10,5))
plt.plot(x, y_g, label='gamma')
plt.plot(x, y_p, label='poisson')
plt.legend()
```
