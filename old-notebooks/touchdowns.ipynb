```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

import lombardi.plot
```

```python
tds = lombardi.plot.player_metric('Tom Brady', 'passing_tds')
num_bins = lombardi.plot.num_bins(tds, 1)
print num_bins
```

```python
data = tds
bin_size = 1
plt.hist(
    [j - bin_size/2 for j in data],
    normed=True,
    bins=lombardi.plot.num_bins(data, bin_size),
    label='hist',
    alpha=0.5,
    width=0.5,
    align='mid',
)
```
