# player distributions

This shows player histograms versus fits.

```python
players = [
    ('Tom Brady', 'passing_yds'),
    ('LeSean McCoy', 'rushing_yds'),
    ('Calvin Johnson', 'receiving_yds'),
]

for player in players:
    data = lombardi.data.player_metric(player[0], player[1], min_year=None, max_year=None)
    lombardi.plot.fit_demo(data, chart_title=player[0], normalize=False)
```

```python
years = [2009, 2010, 2011, 2012, 2013, 2014, 2015]

plt.figure(figsize=(20,10))
for year in years:
    data = lombardi.data.player_metric(
        player='Tom Brady',
        metric='passing_yds',
        min_year=year,
        max_year=year,
    )
    x = np.linspace(0, lombardi.data.metric_range('passing_yds'), 1000)
    lombardi.plot.add_kde(data, x, label=year)
plt.legend()
```
