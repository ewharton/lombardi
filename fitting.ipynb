```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy.stats import rv_discrete
%matplotlib inline

scale = 100
size = 1000

x = np.linspace(-5*scale,5*scale,size)

# generate random normally-distributed data
data = norm.rvs(size=size)
data = [a*scale for a in data]

# fit data to curve
mu, std = norm.fit(data)
y1 = norm.pdf(x, mu, std)
tot = sum(y1)
y1 = [a/tot for a in y1]

# generate random variables from a given fit
custom = rv_discrete(name='custom', values=(x,y1))
data2 = custom.rvs(size=size)

mu2, std2 = norm.fit(data2)
y2 = norm.pdf(x, mu2, std2)
tot2 = sum(y2)
y2 = [a/tot2 for a in y2]

# plot
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gamma
from scipy.stats import gaussian_kde
from scipy.stats import rv_discrete

scale = 100
size = 1000

x = np.linspace(-5*scale,5*scale,size)

# generate random gamma-distributed data
data = gamma.rvs(2, size=size)
data = [a*scale for a in data]

# fit data to curve
kde = gaussian_kde(data)
y1 = kde(x)
tot = sum(y1)
y1 = [a/tot for a in y1]

# generate random variables from a given fit
custom = rv_discrete(name='custom', values=(x,y1))
data2 = custom.rvs(size=size)

a,l,b = gamma.fit(data2)
rv = gamma(a,l,b)
y3 = rv.pdf(x)
tot = sum(y3)
y3 = [a/tot for a in y3]


# plot
plt.figure()
plt.plot(x,y1)
plt.plot(x,y3)
```
